```python
import pandas as pd
```

```python
#path for preprocessed csv
data = pd.read_csv('../../preprocessed.csv')
```

```python
data
```

```python
# remove stopwords from text and summary columns
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
# function to remove stopwords
def remove_stopwords(text):
    text = ' '.join([word for word in text.split() if word not in (stop)])
    return text

#apply remove_stopwords function to text and summary columns
data['text'] = data['text'].apply(lambda x: remove_stopwords(x))

data['summary'] = data['summary'].apply(lambda x: remove_stopwords(x))

data
```

```python
# clean text and summary columns

import re

def clean_text(text):
    # remove numbers
    text = re.sub(r'\([^)]*\)', '', text)
    # remove punctuations and special characters
    text = re.sub('"','', text)
    
    text = ' '.join(text.split())
    return text

data['text'] = data['text'].apply(lambda x: clean_text(x))

data['summary'] = data['summary'].apply(lambda x: clean_text(x))

data
```

```python
#save cleaned data to csv

data.to_csv('data.csv', index=False)
```
